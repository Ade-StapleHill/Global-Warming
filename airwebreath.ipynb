```python
import numpy, pandas
```

```python
df = pandas.read_csv('AirWeBreathe.csv')
pandas.set_option('display.max_rows', 1000)
display(df)
```

```python
i=4 #initial index of measurment data
f=len(df) #final index of measurment data
sf=df[i:f].dropna() #dropping any values with NaN value
sf.reset_index(drop=True, inplace=True) #reindexing dataframe
display(sf)
```

```python
sf.head(3)
```

```python
sf.tail(3)
```

```python
London = sf.iloc[:,3].to_numpy('float')
print(London)
```

```python
Bristol = sf.iloc[:,1].to_numpy('float')
print(Bristol)
```

```python
dates = sf.iloc[:,0]
import datetime as dt
fmt_dates = [dt.datetime.strptime(d,'%d/%m/%Y').date() for d in dates]
print(fmt_dates)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(fmt_dates, London, label='London')
plt.xlabel('Date')
plt.ylabel('Concentraition (units)')
plt.gcf().autofmt_xdate()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(fmt_dates, Bristol, label='Bristol')
plt.xlabel('Date')
plt.ylabel('Concentraition (units)')
plt.gcf().autofmt_xdate()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(fmt_dates, London, label='London')
ax.plot(fmt_dates, Bristol, label='Bristol')
plt.xlabel('Date')
plt.ylabel('Concentraition (units)')
plt.gcf().autofmt_xdate()
plt.legend()
plt.show()
```

```python
London_prelock = London[0:51]
London_postlock = London[51:len(London)]
print(London_prelock)
print(London_postlock)
```

```python
London_prelock_mean=numpy.mean(London_prelock)
London_postlock_mean=numpy.mean(London_postlock)
London_prelock_std=numpy.std(London_prelock)
London_postlock_std=numpy.std(London_postlock)
print('London_prelock:', London_prelock_mean,London_prelock_std)
print('London_postlock:', London_postlock_mean,London_postlock_std)
```

```python
fig=plt.figure()
ax = fig.add_subplot(111)
ax.bar('London_prelock', London_prelock_mean,yerr=London_prelock_std)
ax.bar('London_postlock', London_postlock_mean,yerr=London_postlock_std)
plt.ylabel('Concentraition (units)')
plt.show()
```
