```python
import matplotlib.pyplot as plt
import numpy, pandas, datetime
```

```python
df = pandas.read_csv('AirWeBreathe_seasonal.csv')
pandas.set_option('display.max_rows', 1000)
##display(df)
```

```python
i=4 #initial index of measurment data
f=len(df) #final index of measurment data
sf=df[i:f].dropna() #dropping any values with NaN value
sf.reset_index(drop=True, inplace=True) #reindexing dataframe
##display(sf)
```

```python
Bristol = sf.iloc[:,1].to_numpy('float')
London = sf.iloc[:,3].to_numpy('float')
print('Bristol:', Bristol)
print('London:', London)
```

```python
London_TwN = London[0:178]
London_TwTw = London[178:len(London)]
print(London_TwN)
print(London_TwTw)
```

```python
London_TwN_mean=numpy.mean(London_TwN)
London_TwTw_mean=numpy.mean(London_TwTw)
print('London 2019:', London_TwN_mean)
print('London 2020:', London_TwTw_mean)
```

```python
London_TwN_add = sum(London_TwN)
London_TwTw_add = sum(London_TwTw)
print('London 2019:', London_TwN_add)
print('London 2020:', London_TwTw_add)
```

```python
fig=plt.figure()
ax = fig.add_subplot(111)
ax.bar('2019', London_TwN_mean)
ax.bar('2020', London_TwTw_mean)
plt.title('Avaage Amount of Co2 found in air per day')
plt.ylabel('Concentraition (units)')
plt.show()
```

```python
fig=plt.figure()
ax = fig.add_subplot(111)
ax.bar('2019', London_TwN_add)
ax.bar('2020', London_TwTw_add)
plt.title('all Co2 realesed in the atmosphere in the year at london')
plt.ylabel('Concentraition (ugm)')
plt.show()
```

```python
#convert dates days
dates = sf.iloc[:,0]
day_name= ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday','Sunday']
fmt_days = [datetime.datetime.strptime(d,'%d/%m/%Y').weekday() for d in dates]
fmt_dates = [datetime.datetime.strptime(d,'%d/%m/%Y').date() for d in dates]
for day in fmt_days:
    print(day_name[day])
print(fmt_dates)
```

```python
from datetime import date
import calendar 

today = date.today()

def findDay(date): 
    d = datetime.datetime.strptime(date, '%d %m %Y').weekday() 
    return (calendar.day_name[d]) 
  
# Driver program 
date = today.strftime("%d/%m/%Y")
weekday = findDay(date.replace('/', ' '))
print(weekday) 
```

```python
# set witch day to highlight
if weekday == 'Monday':
    explode = [0.1, 0, 0, 0, 0, 0, 0]
elif weekday == 'Tuesday':
    explode = [0, 0.1, 0, 0, 0, 0, 0]
elif weekday == 'Wednesday':
    explode = [0, 0, 0.1, 0, 0, 0, 0]
elif weekday == 'Thursday':
    explode = [0, 0, 0, 0.1, 0, 0, 0]
elif weekday == 'Friday':
    explode = [0, 0, 0, 0, 0.1, 0, 0]
elif weekday == 'Saturday':
    explode = [0, 0, 0, 0, 0, 0.1, 0]
else:
    explode = [0, 0, 0, 0, 0, 0, 0.1]
```

```python
#find mean of concentraition from all the mondays in each month, then, tuesdays....
Values = sf.iloc[:,1]
Monday = []
Tuesday = []
Wednesday = []
Thursday = []
Friday = []
Saturday = []
Sunday = []
for day in fmt_days:
    if day == 'Monday':
        Monday.append(Values[fmt_days.index(day)])
    elif day == 'Tuesday':
        Tuesday.append(Values[fmt_days.index(day)])
    elif day == 'Wednesday':
        Wednesday.append(Values[fmt_days.index(day)])
    elif day == 'Thursday':
        Thursday.append(Values[fmt_days.index(day)])
    elif day == 'Friday':
        Friday.append(Values[fmt_days.index(day)])
    elif day == 'Saturday':
        Saturday.append(Values[fmt_days.index(day)])
    else:
        Sunday.append(Values[fmt_days.index(day)])
print(Monday)
```

```python
#Calculat the mean avages
Monday_Mean = numpy.mean(Monday)
Tuesday_Mean = numpy.mean(Tuesday)
Wednesday_Mean = numpy.mean(Wednesday)
Thursday_Mean = numpy.mean(Thursday)
Friday_Mean = numpy.mean(Friday)
Saturday_Mean = numpy.mean(Saturday)
Sunday_Mean = numpy.mean(Sunday)
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
days = 'Monday', 'Tuesday', 'wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
### sizes = [457, 698, 584, 896, 776, 334 ,984] #fake didgits 
sizes = [Monday_Mean, Tuesday_Mean, Wednesday_Mean, Thursday_Mean, Friday_Mean, Saturday_Mean, Sunday_Mean]
#highlight day today

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=days, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
